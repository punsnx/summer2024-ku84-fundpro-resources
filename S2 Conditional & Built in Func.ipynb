#### วิธีการใช้ if else ง่ายๆ
- if
- elif
- elif
- else

```python
x = int(input())
if x > 0:
  print(f"X is positive number ({x})")
elif x < 0:
  print(f"X is negative number ({x})")
else:
  print(f"X is 0")
```

#### การเขียน if else ย่อๆ

```python
#if else shorthand

A = 10
print("A is 10") if A == 10 else print("A is not 10")
```

#### การใช้ Try / Except
Try Except มีเอาไว้เพื่อไม่ให้โปรแกรมของเราทำงานผิดพลาด\
เช่น การรับ input ที่มี type ไม่ตรงกันกับที่เราต้องการรับ

```python
#without Try except
x = int(input())
print(x)
```

```python
#with Try except

try:
  x = int(input())
except:
  print("ERROR")
```

#### Min max functions
- ใช้หาค่า min หรือ max ระหว่างข้อมูล หรือชุดข้อมูลหลายตัว

```python
#min max functions

A = 10
B = 8

min_value = min(A,B)
max_value = max(A,B)

print(f"minimum value between A and B is {min_value}")
print(f"maximum value between A and B is {max_value}")
```

```python
#min max (a lot)

theMinimum = min(3,4,5,1,0,-9,10,4,-8,14)
theMaximum = max(3,4,5,1,0,-9,10,4,-8,14)

print(f"min is {theMinimum} and max is {theMaximum}")
```

#### Built in functions อื่นๆที่ใช้บ่อยๆ
- Pow(x,y) คือ x ยกกำลัง y มีค่าเท่ากันกับการใช้ x**y
- abs(x) คืนค่าสัมบูรณ์ของ x

```python
#pow(x,y) => x**y

x = 9
y = pow(x,2)

print(f"The square if {x} is {y}")
```

```python
#abs

x = -55
y = abs(x)

print(f"|{x}| is {y}")
```

#### math functions
- math.sqrt(x) คืนค่ารากที่สองของ x
- math.factorial(x) คือค่า x!

```python
import math

x = 81
square_root = math.sqrt(x)

print(f"Square root of x is {square_root}")
```

```python
import math

x = 4
factorial_of_four = math.factorial(x)

print(f"{x}! = {factorial_of_four}")
```
