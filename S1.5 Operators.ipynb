### Bitwise operator

- Bitwise OR : |
- Bitwise AND : &
- Bitwise XOR : ^
- Bitwise not : ~
- Bitwise lshift : <<
- Bitwise rshift : >>

Assign binary to variable

```python
x = int("101",base=2)
print(x)
x = 0b101
print(x)
print(bin(x))
```

Ex.1 Bitwise OR

```python
x = 0b0101
y = 0b1001
z = x | y
print(bin(z))
```

Ex.2 Bitwise AND

```python
x = 0b1101
y = 0b0110
z = x & y
print(bin(z))
```

Ex.3 Bitwise XOR

```python
x = 0b1001
y = 0b0010
z = x ^ y
print(z)
```

Ex.4 Bitwise not
- previously we mention that (bitwise not) is not the same as (logical not) because it flips the bits.

```python
x = 0b0101
print(x,bin(x))
x = ~x
print(x,bin(x))
#two’s complement representation เรียนใน intro com.
```

Ex.5 Bitwise lshift

```python
x = 0b10110
print(x,bin(x))
x = x << 1
print(x,bin(x))
```

Ex.6 Bitwise rshift

```python
x = 0b10110
print(x,bin(x))
x = x >> 1
print(x,bin(x))
```
